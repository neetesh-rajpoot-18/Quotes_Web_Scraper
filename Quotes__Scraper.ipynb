```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
os.makedirs("data", exist_ok=True)
os.makedirs("output", exist_ok=True)
print("Folders ready!")
```

```python
def scrape_quotes():
    url = "http://quotes.toscrape.com"
    quotes = []

    while url:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, "html.parser")

        for q in soup.find_all("div", class_="quote"):
            text = q.find("span", class_="text").get_text()
            author = q.find("small", class_="author").get_text()
            tags = [t.get_text() for t in q.find_all("a", class_="tag")]
            quotes.append({"quote": text, "author": author, "tags": ", ".join(tags)})

        next_btn = soup.find("li", class_="next")
        url = "http://quotes.toscrape.com" + next_btn.find("a")["href"] if next_btn else None

    print(f"Total quotes scraped: {len(quotes)}")
    return quotes
```

```python
def save_to_csv(quotes):
    df = pd.DataFrame(quotes)
    df.to_csv("data/quotes.csv", index=False)
    print("Saved to data/quotes.csv")
    return df
```

```python
def make_chart(df):
    top_authors = df["author"].value_counts().head(10)

    plt.figure(figsize=(10, 6))
    top_authors.plot(kind="bar", color="steelblue")
    plt.title("Top 10 Authors by Quote Count")
    plt.xlabel("Author")
    plt.ylabel("Number of Quotes")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    path = "output/top_authors.png"
    plt.savefig(path, dpi=150)
    plt.show()
    print(f"Chart saved → {path}")
```

```python
def main():
    print("=" * 45)
    print("  QUOTES SCRAPER")
    print("=" * 45 + "\n")

    quotes = scrape_quotes()
    df = save_to_csv(quotes)

    print("\nPreview of scraped data:")
    print(df.head(5).to_string(index=False))

    make_chart(df)

    print("\n✓ Done!")
    print("  → Open data/quotes.csv to see all quotes")
    print("  → Open output/top_authors.png to see the chart")

main()
```
